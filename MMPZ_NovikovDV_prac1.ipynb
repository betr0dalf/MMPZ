## Задание 1


Блок-схема решения задачи нахождения корней квадратного уравнения



Написать программу для решения квадратного уравнение вида ax**2 + bx + c = 0 по приведенной выше блок-схеме (коэффициенты a, b, c вводятся вручную).

```python
def solve1(a, b, c):
    if a != 0:
        D = b * b - 4 * a * c
        if D >= 0:
            X1 = (-b + D ** 0.5) / (2 * a)
            X2 = (-b - D ** 0.5) / (2 * a)
            print('X1 = ', X1, "; X2 = ", X2)
            return
        else:
            print('Решение находится в области комплексных чисел')
            return
    elif b != 0:
        print('X = ', -c / b)
        return
    elif c != 0:
        print('Нет решений')
        return
    else:
        print('X - любое число')

print('Задание 1\n')
a = float(input('Введите a: '))
b = float(input('Введите b: '))
c = float(input('Введите c: '))

solve1(a, b, c)
```


## Задание 2
Определить программными средствами, является ли заданное число а степенью числа 3 (число а вводится вручную). То есть является ли число а равным 3**n. Желательно не прибегать к использованию циклов.

```python
def solve2(a):
    if a <= 0:
        return False
    while a % 3 == 0:
        a //= 3
    return a == 1


print('Задание 2\n')
a = float(input('Введите a: '))

if solve2(a):
    print(f"Число {a} является степенью числа 3")
else:
    print(f"Число {a} не является степенью числа 3")
```
