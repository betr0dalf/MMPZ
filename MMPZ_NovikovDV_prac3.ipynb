Используя формулу (1) написать программу для расчета параметров реактивного самолета. Необходимо вручную задать три любых параметра из формулы (площадь крыла, скорость, угол отклонения, масса), по которым будет определен 4-й параметр.

```python
from math import pi


print('Задание 1\n')

print('Введите 1, чтобы найти площадь крыла')
print('Введите 2, чтобы найти скорость')
print('Введите 3, чтобы найти угол отклонения')
print('Введите 4, чтобы найти массу\n')

checker = int(input('Введите число: '))

while checker not in (1, 2, 3, 4):
    print('Введено неверное значение!')
    checker = int(input('Введите число от 1 до 4 включительно: '))

if checker == 1:
    v = float(input('Введите скорость: '))
    a = float(input('Введите угол отклонения: '))
    M = float(input('Введите массу: '))

    S = (M * 10) / (v ** 2 * (a * pi / 180) ** 2)

    print('\nПлощадь крыла S =', round(S, 3))

elif checker == 2:
    S = float(input('Введите площадь крыла: '))
    a = float(input('Введите угол отклонения: '))
    M = float(input('Введите массу: '))

    v = (((M * 10) / S) / ((a * pi / 180) ** 2)) ** 0.5

    print('\nСкорость v =', round(v, 3))

elif checker == 3:
    S = float(input('Введите площадь крыла: '))
    v = float(input('Введите скорость: '))
    M = float(input('Введите массу: '))

    a = ((((M * 10) / S) / (v ** 2)) ** 0.5) * 180 / pi

    print('\nУгол отклонения a =', round(a, 3))

elif checker == 4:
    S = float(input('Введите площадь крыла: '))
    v = float(input('Введите скорость: '))
    a = float(input('Введите угол отклонения: '))

    M = S * v ** 2 * (a * pi / 180) ** 2 / 10

    print('\nМасса M =', round(M, 3))


# (площадь крыла l * d, скорость v, угол отклонения a, масса M)
```
